```python
import pandas as pd 
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('./BostonHousing(2).xlsx', header=0)

X = df.iloc[:, range(13)]
y = df['CAT. MEDV']
dy = df.iloc[:, 14]

#train and test split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, random_state=0)
```

```python
#depth of 2

cf = DecisionTreeClassifier(max_depth = 2)
t1 = cf.fit(X_train, Y_train)
score1 = cf.score(X_test, Y_test)
tree.plot_tree(t1)
print(score1)  
```

```python
#depth =3
cf = DecisionTreeClassifier(max_depth = 3)
t2 = cf.fit(X_train, Y_train)
tree.plot_tree(t2)
score2 = cf.score(X_test, Y_test)
print(score2)
```

```python
accuracy=[]
max_depth_list=[2,3,4,5,6,7,8,9]
for n in max_depth_list:     
    cf = DecisionTreeClassifier(max_depth = n)   
    cf.fit(X_train, Y_train)   
    score = cf.score(X_test, Y_test)
    accuracy.append(score)
    print('Tree with {} max_depth has an accuracy of {}'.format(n, score))
    
```

```python
#brief analysis

#creating domain for max_depth for plot
max_depth = np.arange(len(max_depth_list)) + 2
plt.plot(max_depth, accuracy)
plt.xlabel('Maximum Depth') # Label x-axis
plt.ylabel('Accuracy') # Label y-axis
plt.title("Max_depth against accuracy")
plt.show()
```
